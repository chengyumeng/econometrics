# 分布形状

### 累积分布函数 cdf

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

df = stats.norm(5, 3)

x = np.linspace(-5, 15, 101)
y = df.cdf(x)


fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
```

### 概率密度函数 pdf

```python
z = df.pdf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, z)
plt.show()
```

### 生存函数 sf=1-cdf

```python
p = df.sf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, p)
plt.show()
```

### 百分点函数 ppf

```python
q = df.ppf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, q)
plt.show()
```

### 逆生存函数 isf 

```python
i = df.isf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, i)
plt.show()
```
